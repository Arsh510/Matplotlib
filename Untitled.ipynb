```python
%matplotlib inline
from matplotlib import pyplot as plt

x = [2,5,7,4,6]
y = [8,6,4,2,5]
x1 = [1,3,5,7,9]
y1 = [3,6,7,9,1]

plt.plot(x,y, label="first")
plt.plot(x1,y1, label="second")
plt.legend()

plt.show()
```

```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
x=np.linspace(-1,1)
```

```python
y=x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```
